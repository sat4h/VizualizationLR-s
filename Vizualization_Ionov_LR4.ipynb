Я сначала сделал для нескольких переменных для теста

```python
# Преобразуем категориальные данные в строковые для удобной визуализации
df['cyl'] = df['cyl'].astype(str)
df['vs'] = df['vs'].astype(str)
df['am'] = df['am'].astype(str)
df['gear'] = df['gear'].astype(str)
df['carb'] = df['carb'].astype(str)

df_wide = df.pivot_table(index='model', columns='cyl', values=['mpg', 'hp', 'wt'], aggfunc='mean')

# Попробуем визуализировать
df_wide.plot(kind='bar', figsize=(10, 6))
plt.title('Сравнение по категориям cyl')
plt.ylabel('Значения')
plt.show()
```

```python
df_wide.info()
```

```python
df_wide
```

```python
# Разделим числовые и категориальные столбцы
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Применим pivot_table только к числовым столбцам
df_wide_numerical = df.pivot_table(index='model', columns='cyl', values=numerical_cols, aggfunc='mean')

# Для категориальных данных оставим их как есть (можно использовать, например, агрегацию по моде)
df_wide_categorical = df[categorical_cols].groupby('model').first()

# Теперь объединим числовые и категориальные данные в одну таблицу
df_wide_all = pd.concat([df_wide_numerical, df_wide_categorical], axis=1)
```

```python
# Пример 1: Столбчатая диаграмма расхода топлива (mpg) по количеству цилиндров
df_wide_all[['mpg_4', 'mpg_6', 'mpg_8']].plot(kind='bar', figsize=(12, 6))
plt.title('Сравнение расхода топлива (mpg) по количеству цилиндров')
plt.ylabel('Расход топлива (mpg)')
plt.xticks(rotation=90)
plt.tight_layout()  # Для лучшего отображения подписей на осях
plt.show()
```

```python
print(df_wide_all.columns)
```

```python
# Преобразуем MultiIndex в строки (('mpg', '4') -> 'mpg_4')
df_wide_all.columns = ['_'.join(map(str, col)) if isinstance(col, tuple) else col for col in df_wide_all.columns]

print(df_wide_all.columns)
```

```python
# Столбчатая диаграмма для расхода топлива (mpg) по количеству цилиндров
df_wide_all[['mpg_4', 'mpg_6', 'mpg_8']].plot(kind='bar', figsize=(12, 6))
plt.title('Сравнение расхода топлива (mpg) по количеству цилиндров')
plt.ylabel('Расход топлива (mpg)')
plt.xticks(rotation=90)
plt.tight_layout()  # Для лучшего отображения подписей на осях
plt.show()
```

```python
# Столбчатая диаграмма для веса (wt) по количеству цилиндров
df_wide_all[['wt_4', 'wt_6', 'wt_8']].plot(kind='bar', figsize=(12, 6))
plt.title('Сравнение веса (wt) по количеству цилиндров')
plt.ylabel('Вес (wt)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

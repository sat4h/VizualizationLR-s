```python
from plotnine import *
```

*1. Гистограмма распределения мощности (hp)*

image.png

```python
ggplot(df, aes(x='hp')) +\
geom_histogram(bins=10, fill='skyblue', color='black') +\
labs(title='Распределение мощности (hp)', x='Мощность', y='Частота')
```

*2. Диаграмма рассеяния: мощность (hp) vs вес (wt)*

image.png

```python
ggplot(df, aes(x='hp', y='wt')) +\
geom_point(color='red', alpha=0.7) +\
labs(title='Мощность vs Вес', x='Мощность (hp)', y='Вес (wt)')
```

*3. Линейный график зависимости мощности от числа цилиндров*

image.png

```python
ggplot(df, aes(x='cyl', y='hp')) + \
stat_summary(fun_y=mean, geom='line', color='blue', linetype='-', size=1) + \
stat_summary(fun_y=mean, geom='point', color='blue', size=3) + \
labs(title='Средняя мощность по числу цилиндров', x='Число цилиндров', y='Средняя мощность (hp)')
```

*4. "Ящик с усами" для распределения расхода топлива (mpg) по числу цилиндров*

image.png

```python
ggplot(df, aes(x='factor(cyl)', y='mpg')) + \
geom_boxplot(fill='skyblue', width=0.3) + \
labs(title='Распределение расхода топлива (mpg) по числу цилиндров', x='Число цилиндров', y='Миль на галлон (mpg)')
```

*8. Столбчатая диаграмма: средний расход топлива (mpg) по числу передач (gear)*

image.png

*9. Линейный график: изменение мощности (hp) по порядку машин*

image.png

```python
ggplot(df, aes(x=df.index, y='hp')) +\
geom_line(color='magenta') +\
geom_point(color='magenta') +\
labs(title='Изменение мощности по порядку машин', x='Порядок машин', y='Мощность (hp)')
```

*10. Тепловая карта корреляций*

image.png

```python
# Выбираем только числовые столбцы
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Вычисляем матрицу корреляций
corr_matrix = df_numeric.corr()

# Преобразуем матрицу корреляций в длинный формат
corr_long = corr_matrix.reset_index().melt(id_vars="index")
corr_long.columns = ['Variable1', 'Variable2', 'Correlation']

ggplot(corr_long, aes(x='Variable1', y='Variable2', fill='Correlation')) +\
geom_tile() +\
scale_fill_gradient2(low="blue", high="red", mid="white", midpoint=0) +\
labs(title='Тепловая карта корреляций', x='', y='')
```

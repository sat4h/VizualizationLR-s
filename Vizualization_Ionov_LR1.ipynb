```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
```

Я взял датасет [mtcars](https://www.kaggle.com/datasets/ruiromanini/mtcars/data). Как можно понять по названю он состоит из данных по машинам.

```python
# Download latest version
path = kagglehub.dataset_download("ruiromanini/mtcars")

print("Path to dataset files:", path)
```

```python
# Чтение данных
df = pd.read_csv(f"{path}/mtcars.csv")
```

```python
df.head(10)
```

Размер датасета: 32 строки (Количество автомобилей) и 11 столбцов.

Теперь можно построить 10 различных графиков. Я буду использовать matplotlib.


---



*1. Гистограмма распределения мощности (hp)*

```python
plt.figure(figsize=(10, 5))
plt.hist(df['hp'], bins=10, color='skyblue', edgecolor='black')
plt.title('Распределение мощности (hp)')
plt.xlabel('Мощность')
plt.ylabel('Частота')
plt.grid(axis='y')
plt.show()
```

*2. Диаграмма рассеяния: мощность (hp) vs вес (wt)*

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['hp'], df['wt'], color='red', alpha=0.7)
plt.title('Мощность vs Вес')
plt.xlabel('Мощность (hp)')
plt.ylabel('Вес (wt)')
plt.grid(True)
plt.show()
```

*3. Линейный график зависимости мощности от числа цилиндров*

```python
mean_hp_by_cyl = df.groupby('cyl')['hp'].mean()
plt.figure(figsize=(10, 5))
plt.plot(mean_hp_by_cyl.index, mean_hp_by_cyl.values, marker='o', linestyle='-', color='blue')
plt.title('Средняя мощность по числу цилиндров')
plt.xlabel('Число цилиндров')
plt.ylabel('Средняя мощность (hp)')
plt.grid()
plt.show()
```

*4. "Ящик с усами" для распределения расхода топлива (mpg) по числу цилиндров*

```python
plt.figure(figsize=(10, 5))
df.boxplot(column='mpg', by='cyl', grid=False, patch_artist=True)
plt.title('Распределение расхода топлива (mpg) по числу цилиндров')
plt.suptitle('')  # Убираем общий заголовок
plt.xlabel('Число цилиндров')
plt.ylabel('Миль на галлон (mpg)')
plt.show()
```

*5. Круговая диаграмма распределения числа машин по числу цилиндров*

```python
cyl_counts = df['cyl'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(cyl_counts, labels=cyl_counts.index, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'pink'])
plt.title('Распределение машин по числу цилиндров')
plt.show()
```

*6. График плотности для распределения веса (wt)*

```python
plt.figure(figsize=(10, 5))
df['wt'].plot(kind='density', color='purple')
plt.title('График плотности веса (wt)')
plt.xlabel('Вес (wt)')
plt.grid()
plt.show()
```

*7. Парные зависимости с использованием `scatter_matrix` (из pandas)*

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df[['mpg', 'hp', 'wt', 'qsec']], figsize=(12, 12), alpha=0.8, diagonal='kde')
plt.suptitle('Парные зависимости')
plt.show()
```

*8. Столбчатая диаграмма: средний расход топлива (mpg) по числу передач (gear)*

```python
mean_mpg_by_gear = df.groupby('gear')['mpg'].mean()
plt.figure(figsize=(10, 5))
plt.bar(mean_mpg_by_gear.index, mean_mpg_by_gear.values, color=['cyan', 'orange', 'lime'])
plt.title('Средний расход топлива по числу передач')
plt.xlabel('Число передач')
plt.ylabel('Средний расход топлива (mpg)')
plt.show()
```

*9. Линейный график: изменение мощности (hp) по порядку машин*

Бесполезный пример Порядок машин/мощности, но для примера сойдет :)

```python
plt.figure(figsize=(10, 5))
plt.plot(df['hp'], marker='o', linestyle='-', color='magenta')
plt.title('Изменение мощности по порядку машин')
plt.xlabel('Порядок машин')
plt.ylabel('Мощность (hp)')
plt.grid()
plt.show()
```

*10. Тепловая карта корреляций*

Перед построением тепловой карты корреляций нужно удалить или игнорировать нечисловые столбцы

Здесь график создается с помощью библиотеки seaborn

```python
# Убираем нечисловые столбцы
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Тепловая карта корреляций
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Тепловая карта корреляций')
plt.show()
```
